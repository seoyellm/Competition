### 방법1) apply(lambda x : )

```python
traincp = train.copy()
def mydef(x):  # x == df
    day_type = 0
    if   x['holiday'] == 0 and x['workingday'] == 0 :
        day_type = 0
    elif x['holiday'] == 0 and x['workingday'] == 1 :
        day_type = 1        
    elif x['holiday'] == 1 and x['workingday'] == 0 :        
        day_type = 2
    return day_type
        
    
traincp["day_type1"] = traincp.apply(lambda df : mydef(df), axis=1)
traincp[['holiday','workingday','day_type1']].value_counts().sort_index()
```

```python
traincp = train.copy()
def mydef(hol, work):
    day_type = 0
    if   hol == 0 and work == 0 :
        day_type = 0
    elif hol == 0 and work == 1 :
        day_type = 1        
    elif hol == 1 and work == 0 :        
        day_type = 2
    return day_type
        
# traincp["day_type1"] = traincp.apply(lambda df : mydef(df), axis=1)    
traincp["day_type2"]   = traincp.apply(lambda df : mydef(df['holiday'], df['workingday']), axis=1)
traincp[['holiday','workingday','day_type2']].value_counts().sort_index()
```

# 학습&평가
- 수식 작성(latex) : https://www.latex4technics.com/?note=CMEL1Z
- https://suboptimal.wiki/explanation/mse/

> latex
*  $ MSE = \frac{\mathrm{1} }{\mathrm{n}} \sum\nolimits_{i=1}^{n} (Y_i - \hat{Y_i})^{2}$

*  $ RMSE = \sqrt{\frac{1}{n}\Sigma_{i=1}^{n}{\Big(\frac{Y_i - \hat{Y_i}}{N}\Big)^2}}$

*  $ RMSLE = \sqrt{\frac{1}{n}\Sigma_{i=1}^{n}{\Big(\frac{log(Y_i+1) - log(\hat{Y_i}+1)}{N}\Big)^2}}$

*  $ \text{MAE}(y, \hat{y}) = \frac{ \sum_{i=0}^{N - 1} |y_i - \hat{y}_i| }{N}$

> score
- mean_squared_error
    - 에러 제곱의 평균(MSE) 
        - 에러에 대해 심한 패널티 부여함 , 작은 차이임에도 큰 오차로 결과를 보여줌, outlier에 민감하게 반응
        - 실제값과 예측값의 차이를 제곱한 값을 보여줌
        - RMSE : MSE에 루트씌움
        - overfitting의 위험이 있음 : loss를 최소화시키는 학습을 하기 때문 --> 모델 불안정
        - test 데이터에 outlier많은 경우는 MSE를 줄여서는 안된다 
            - MSE가 점점 낮아진다는 것은 overfitting에 가까워지고 있다는 뜻
- mean_absolute_error
    - 에러의 절대값 평균(MAE)
        - outlier에 가장 민감하지 않음
        - RMSE와 비슷, 그러나 MAE가 수치 더 작음(RMSE > MAE)
        - RMSE와 MAE 비교
            - 모든 오차에대해 가중치가 동일  --> 가장 쓸데없음
            - RMSE는 오차에대해 가중치가 다름(제곱을하기 때문에) --> 오차에 대해 패널티 부여
            - RMSE가 MAE와 수치가 거의 동일하다는 것은 매우 잘 맞췄다는 뜻 --> outlier의 근사치로 맞췄다는 뜻
- 회귀에서는 accuracy_score 계산하지 않음 --> 정확히 일치하는 값을 찾는 것은 거의 불가능하기 때문
- bike에서는 통상적으로 얼마나 빌리는지를 알아내야하기 때문에 outlier에 패널티를 크게 주어야함 --> MSE 사용해야 함
    - RMSLE_score : RMSE에 log값을 씌움
    - 
    - log + 1 : +1안하면 0에 수렴하기 때문 --> underflow

```python
y_list = [ train['regcount'], train['casual'], train['registered'] ]
y_col  = [ 'regcount', 'casual', 'registered'] 
X = train.drop(['regcount','casual','registered'], axis=1)
for i, y in enumerate(y_list) :
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
    dtr = DecisionTreeRegressor(random_state=0)
    dtr.fit(X_train, y_train)
    y_pred = dtr.predict(X_test)
    score = mean_squared_error(y_test, y_pred)
    print( y_col[i] )
    print("MSE : ", score) 
    print("RMSE: ", np.sqrt(score))
```

```python
model_list = [ ("DTR"   , DecisionTreeRegressor(random_state=0)),
               ("LR"    , LinearRegression()                   ),
               ("ABOOST", AdaBoostRegressor()                  ),
               ("XGB"   , XGBRegressor()                       ),  
               ("LGBM"  , LGBMRegressor()                      ),
               ("VR"    , VotingRegressor([("XGB", XGBRegressor()), ("LGBM", LGBMRegressor())]) ) 
             ]
y_list = [ train['regcount'], train['casual'], train['registered'] ]
y_col  = [ 'regcount', 'casual', 'registered'] 
X = train.drop(['regcount','casual','registered'], axis=1)

score_list = []
for tpl in model_list :
    print( tpl[0] ) 
    for i, y in enumerate(y_list) :
        model = tpl[1]
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        mse_score = mean_squared_error(y_test, y_pred)
        
        print( y_col[i] )
        print("MSE : ", mse_score) 
        print("RMSE: ", np.sqrt(mse_score)) 
        print("-"*30)
        score_list.append([tpl[0], y_col[i], mse_score, np.sqrt(mse_score) ] )
```
